```
sh = shuffler(arts)
```

```
sh_d1=sh["Articles"]
sh_d2=sh["Categories"]
```

```
ts1, tr1, ts2, tr2 = model_selection.train_test_split(sh_d1, sh_d2, test_size=0.2, random_state=4)
```

```
vectorizer = CountVectorizer()
```

```
tr1_t = vectorizer.fit_transform(tr1)
print( tr1_t )
```

```
print( vectorizer.vocabulary_ )
```

```
tr2 = tr2.astype('str')
```

```
mlpClass = MLPClassifier(hidden_layer_sizes=(100, 100, 100), max_iter = 200)
svcClass = SVC(kernel = 'linear', gamma = 'auto', C = 0.5)
dstClass = DecisionTreeClassifier(criterion="entropy")
```

```
 mlpClass.fit(tr1_t, tr2)
 svcClass.fit(tr1_t, tr2)
 dstClass.fit(tr1_t, tr2)
```

```
def Av_P_R(val2, predAlgo):
  from sklearn.metrics import recall_score
  from sklearn.metrics import confusion_matrix
  print('Classification Report: \n', classification_report(val2, predAlgo)),
  print('Confusion Matrix: \n', confusion_matrix(val2, predAlgo)), 
  print('Recall Score: \n', recall_score(val2, predAlgo, average='micro'))
```

```
def SVML_Algo(val1,val2,algo):
  predAlgo = algo.predict(val1)
  return Av_P_R(val2, predAlgo)
```

```
crv_t = vectorizer.transform(ts1)
```

```
print(SVML_Algo(crv_t, ts2, mlpClass))
```

```
print(SVML_Algo(crv_t, ts2, svcClass))
```

```
print(SVML_Algo(crv_t, ts2, dstClass))
```
