```
cv1 = cv.iloc[:,1:]
cv2 = cv.iloc[:,0]
ts1 = ts.iloc[:,1:]
ts2 = ts.iloc[:,0]
tr1 = tr.iloc[:,1:]
tr2 = tr.iloc[:,0]
sh1 = sh.iloc[:,1:]
sh2 = sh.iloc[:,0]
```

```
def SVML_Algo(val1,val2,algo):
  predAlgo = algo.predict(val1)
  return classification_report(val2, predAlgo)
```

```
def trDataF(nn, svm, dt):
  nn.fit(tr1, tr2)
  svm.fit(tr1, tr2)
  dt.fit(tr1, tr2)
```

```
mlpClass = MLPClassifier(hidden_layer_sizes=(100, 100, 100), max_iter = 500)
svcClass = SVC()
dstClass = DecisionTreeClassifier(criterion="entropy")
```

```
trDataF(mlpClass, svcClass, dstClass)
```

```
k = model_selection.KFold(n_splits=10, random_state=None)
```

```
def mdSelect(algo):
  mdl = model_selection.cross_val_score(algo, sh1, sh2, cv=k)
  return mdl.mean()
```

```
print("Accuracy: " , mdSelect(MLPClassifier()))
```

```
print("Accuracy: " , mdSelect(SVC()))
```
